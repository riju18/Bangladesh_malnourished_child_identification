```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Child_Malnutrition_Dataset.csv')
df.head()
```

```python
df = df.drop(['PSU','Wealth Index','BORD','HAZ','WAZ','WHZ','Children_ever_born'],axis=1)
```

```python
df.info()
```

```python
df.head()
```

```python
# convert diar col str to int
# -----------------------------

diarr = df.iloc[:,6].values
diarrr = []
for i in diarr:
    if i.strip():
        n =int(i)
        diarrr.append(n)

diarh = pd.DataFrame({'dia':diarrr})
df['diar'] = diarh
```

```python
# convert fever col str to int
# -----------------------------

fev = df.iloc[:,7].values
fevr = []
for i in fev:
    if i.strip():
        n =int(i)
        fevr.append(n)

feverr = pd.DataFrame({'fever':fevr})
df['fever'] = feverr
```

```python
# convert mother_bmi col str to float
# ----------------------------------

moth_bmi = df.iloc[:,9].values
mothr_bmi = []
for i in moth_bmi:
    if i.strip():
        n =float(i)
        mothr_bmi.append(n)

bmi = pd.DataFrame({'bmi':mothr_bmi})
df['Mother_BMI'] = bmi
```

```python
# convert father_edu col str to int
# ---------------------------------

fatherEdu = df.iloc[:,13].values
father_Edu = []
for i in fatherEdu:
    if i.strip():
        n =float(i)
        father_Edu.append(n)

father_edu = pd.DataFrame({'Father_Edu':father_Edu})
df['Father_Edu'] = father_edu
```

```python
# convert working_mother col str to int
# ---------------------------------

working_mot = df.iloc[:,16].values
working_Mot = []
for i in working_mot:
    if i.strip():
        n =float(i)
        working_Mot.append(n)

currently_work = pd.DataFrame({'currently_working_mot':working_Mot})
df['currently_working_mot'] = currently_work
```
